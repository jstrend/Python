## First Steps Towards Programming

```python
# Fibonacci series:
# the sum of two elements defines the next
a, b = 0, 1
while a < 10:
     print(a)
     a, b = b, a+b
```

###  < (less than), > (greater than), == (equal to), <= (less than or equal to), >= (greater than or equal to) and != (not equal to)

```python
i = 256*256
print('The value of i is', i)
```

```python
a, b = 0, 1
while a < 1000:
    print(a, end=',')
    a, b = b, a+b
```
