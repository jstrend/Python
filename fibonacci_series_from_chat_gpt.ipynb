```python
def fibonacci(n):
  if n <= 0:
    print("Incorrect input")
  elif n == 1:
    return 0
  elif n == 2:
    return 1
  else:
    fib = [0,1]
    for i in range(2,n):
      fib.append(fib[i-1]+fib[i-2])
    return fib

print(fibonacci(10))
```
